AAPL Historical Prices

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Creating a Dataframe

```python
dataframe = pd.read_csv("AAPLHistoricalData.csv")
print(dataframe)
```

```python
dataframe.info()
```

```python
dataframe.head()
```

```python
sns.countplot(x = dataframe['Date'])
```

```python
dataframe['Date'] = pd.to_datetime(dataframe['Date'])
grouped_data = dataframe.groupby('Price')['Date'].count()
result = pd.DataFrame({'Date': grouped_data})
plt.plot(result, c="green", marker="o")
plt.xlabel("Price of AAPL", c="red", size=20)
plt.ylabel("Date", c="red", size=20)
```

```python
pivot_table = dataframe.pivot_table(index = 'High', columns='Low', aggfunc='size', fill_value=0)
sns.heatmap(pivot_table, annot = True, cmap="plasma", fmt='d')
plt.title("HeatMap")
plt.xlabel("Low Of AAPL", c ="Red", size =20)
plt.ylabel("High Of AAPL", c ="Red", size = 20)
plt.show()
```
